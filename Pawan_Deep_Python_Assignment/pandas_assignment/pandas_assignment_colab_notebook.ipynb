```python
import pandas as pd
import requests
import json
```

```python
#making dataframe 1
r1=requests.get("https://reqres.in/api/users")
json=r1.json()
df1 = pd.DataFrame(json['data'])
df1.drop('avatar',axis='columns', inplace=True)
df1
```

```python
#making dataframe 2
r2=requests.get("https://reqres.in/api/users?page=2")
json=r2.json()
df2= pd.DataFrame(json['data'])
df2.drop('avatar',axis='columns', inplace=True)
df2
```

```python
#compare both the data frame.
df1.compare(df2,keep_equal=True)
```

```python
#append both the dataframe and sort them with the column value first_name
df3 = df1.append(df2, ignore_index=True)
df4=df3.sort_values('first_name')
df4
```

```python
#making final dataframe consisting of IDs: 4-10 inclusive.
final = df3[(df3['id'] >= 4) & (df3['id']<=10)]
final
```
